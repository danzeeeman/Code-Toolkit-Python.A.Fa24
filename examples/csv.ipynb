```python
import csv
import random
# Path to the CSV file
file_path = 'Koyaa json captions - Sheet1.csv'

prompts = []
# Open the CSV file
with open(file_path, mode='r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    
    # Loop over each row in the CSV
    for row in csv_reader:
        # Do something with each row
        if 'credits' not in row[1] and 'people' in row[1]:
            if random.randrange(0, 100000) < 1500:
                prompts.append(row[2])
                
deforum_string = ""
frame_num = 0
for prompt in prompts:
    # cleaned = prompt.replace('"', '')
    deforum_string = f"{deforum_string}{frame_num}:\" koyaanisqatsi {prompt}\",\n"
    frame_num += 45

print(deforum_string)
```
